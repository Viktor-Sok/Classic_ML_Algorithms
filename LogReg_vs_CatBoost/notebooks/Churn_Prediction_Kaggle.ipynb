# Метрика

Перед решением любой задачи важно понимать, как будет оцениваться ваше решение. В данном случае мы используем стандартную для задачи классификации метрику ROC-AUC. Ее можно вычислить, используя только предсказанные вероятности и истинные классы без конкретного порога классификации + она раотает даже если классы в данных сильно несбалансированны (примеров одного класса в десятки раз больше примеров длугого). Именно поэтому она очень удобна для соревнований.

Посчитать ее легко:


```python
from sklearn.metrics import roc_auc_score

y_true = [
    0,
    1,
    1,
    0,
    1
]

y_predictions = [
    0.1,
    0.9,
    0.4,
    0.6,
    0.61
]

roc_auc_score(y_true, y_predictions)
```

```python
data = pd.read_csv('./train.csv')
```

```python
# Для вашего удобства списки с именами разных колонок

# Числовые признаки
num_cols = [
    'ClientPeriod',
    'MonthlySpending',
    'TotalSpent'
]

# Категориальные признаки
cat_cols = [
    'Sex',
    'IsSeniorCitizen',
    'HasPartner',
    'HasChild',
    'HasPhoneService',
    'HasMultiplePhoneNumbers',
    'HasInternetService',
    'HasOnlineSecurityService',
    'HasOnlineBackup',
    'HasDeviceProtection',
    'HasTechSupportAccess',
    'HasOnlineTV',
    'HasMovieSubscription',
    'HasContractPhone',
    'IsBillingPaperless',
    'PaymentMethod'
]

feature_cols = num_cols + cat_cols
target_col = 'Churn'
```

```python
# посмотрим на случайные строки
data.sample(5)
```

```python
#  видно что поля TotalSpent имеет неправильный тип данных
data.info()
```

```python
# в колонке TotalSpent есть пустые строки вида " ", за меним их на np.nan  перед конвертацией к численному типу данных
data["TotalSpent"] = data["TotalSpent"].apply(lambda val: val if val != " " else np.nan)
data["TotalSpent"] = pd.to_numeric(data["TotalSpent"])
```

```python
# посмотрим на пропуски в численных колонках
data.isna().sum()
```

```python
# статистика числовых данных
data[num_cols].describe()
#data.describe(include=["object"])
```

```python
# статистика категориальных данных
data[cat_cols].describe(include = ["object"])
```

```python
# Так как пропусков немного по сравнению с размером датасета (9 vs 5282) просто удалим соответствующие строки
data.dropna(inplace=True)
data.shape
```

```python
# проверим на дубликаты
data[data.duplicated()].shape
```

```python
# избавимся от дубликатов
data.drop_duplicates(inplace=True)
```

```python
# посмотрим на категории в категориальных данных на предмет мусора в категориях
# Видим странное значение No internet service во многих категориальных признаках, а также No phone service
# кажется что онидублируют значения признаков HasPhoneService и HasInternetService разберёмся с этим подробнее
for col_name, col in data[cat_cols].items():
  print(f"{col_name}: {col.unique()}")
```

```python
# Из графика мы видим, что values 'No phone service' и 'No internet service' из колонок 'HasPhoneService' и 'HasInternetService' дублируются в других колонках
#  а значит не несут новой информации и их можно заменить на значение 'No'  в этих колонках
# Heatmap of missing values
plt.figure(figsize=(12,6))
dublicate_values = (data[cat_cols] == "No internet service") | (data[cat_cols] == "No phone service")
origin_values = data[cat_cols] == "Fill in with False values"
origin_values["HasPhoneService"] = data["HasPhoneService"] == "No"
origin_values["HasInternetService"] = data["HasInternetService"] == "No"
sns.heatmap((dublicate_values | origin_values).T, cmap='summer')
plt.title("Heatmap of 'No internet service value'")
```

```python
# выделим имена колонок где нам нужно выполнить замену значений
dublicate_cols = []
for name, val in (dublicate_values.sum() > 0).items():
  if val:
    dublicate_cols.append(name)
dublicate_cols
```

```python
# заменяем 'No phone service' и 'No internet service' в выбранных выше колонках знаечнием "No"
for name in dublicate_cols:
  data[name] = data[name].replace({"No phone service" : "No", "No internet service" : "No"})
```

```python
# проверим результат замена категориальных значений
for col_name, col in data[cat_cols].items():
  print(f"{col_name}: {col.unique()}")
```

```python
from sklearn.impute import SimpleImputer

def preprocess_test_data(data, cat_cols, num_cols) -> pd.DataFrame:
  """
  Combines data cleaning, filling missing values and feature engeeniring
  """
  # Check for missing values
  assert data.isna().sum().sum() == 0, "Some values are missing"

  # change datatype of TotalSpent
  data["TotalSpent"] = data["TotalSpent"].apply(lambda val: val if val != " " else np.nan)
  data["TotalSpent"] = pd.to_numeric(data["TotalSpent"])

  # fill in missing values with median
  imp = SimpleImputer(missing_values=np.nan, strategy='median')
  data["TotalSpent"] = imp.fit_transform(data[["TotalSpent"]])

  # correcting some categorical variables
  dublicate_values = (data[cat_cols] == "No internet service") | (data[cat_cols] == "No phone service")
  for name, val in (dublicate_values.sum() > 0).items():
    if val:
      data[name] = data[name].replace({"No phone service" : "No", "No internet service" : "No"})

  # delete useless features
  useless_features = ["Sex", "HasMultiplePhoneNumbers", "HasPhoneService",  "HasOnlineTV", "HasDeviceProtection"]
  cat_cols = list(set(cat_cols) - set(useless_features))
  data.drop(useless_features, inplace=True, axis = 1)

  # creating new features
  data['MonthLimit'] = np.nan
  data.loc[data['MonthlySpending']<=60.0,'MonthLimit'] = 0 # клиент тратящий менее 60 в месяц с меньшей веройстностью уйдёт
  data.loc[data['MonthlySpending'] > 60.0,'MonthLimit'] = 1
  data['MonthLimit'] = data['MonthLimit'].astype('int')

  data['ClientPeriodLimit'] = np.nan
  data.loc[data['MonthlySpending']<=30.0,'ClientPeriodLimit'] = 0
  data.loc[data['MonthlySpending'] > 30.0,'ClientPeriodLimit'] = 1
  data['ClientPeriodLimit'] = data['ClientPeriodLimit'].astype('int')

  # log transform
  data["TotalSpentLog"] = data["TotalSpent"].apply(lambda x: np.log(x))
  data.drop("TotalSpent", axis = 1, inplace=True)
  # add new features to lists
  cat_cols += ["MonthLimit", "ClientPeriodLimit"]
  num_cols = [col for col in num_cols if col != "TotalSpent"]
  num_cols += ["TotalSpentLog"]

  return (data, cat_cols, num_cols)
```

```python
data_test, catCols, numCols = preprocess_test_data(data_test, cat_cols, num_cols)
```
